```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('C:/Users/Han/Desktop/Statistics/Python/Dataset/Seoul_Happiness_2014_train.csv')
```

```python
part = data.loc[:, 'Q34':'Q42C2']
part.head()
```

**Q34~Q36 그대로 사용**

### Q36A1, Q36A2 통합(SUM)

```python
part['Q36_1'] = part.loc[:, 'Q36A1'] + part.loc[:, 'Q36A2']
part.drop(['Q36A1', 'Q36A2'], axis=1, inplace=True)
```

**Q37 그대로 사용**

### Q38~Q39 범주형으로 변경

```python
sum(part.loc[:, 'Q38'] == 6) + sum(part.loc[:, 'Q38'] == 98) # 6번(기타) 응답한 경우 0명 + 98로 응답한 경우 7명
```

```python
part = part.drop(part[part['Q38'] == 98].index)
```

```python
part[['Q38_2', 'Q38_3', 'Q38_4', 'Q38_5']] = pd.get_dummies(part['Q38'], prefix = 'Q38', drop_first=True)
part.drop(['Q38'], axis=1, inplace=True)
```

```python
sum(part.loc[:, 'Q39'] == 5) + sum(part.loc[:, 'Q39'] == 98) # 5번(기타) 응답한 경우 8명 + 98로 응답한 경우 2명
```

```python
part = part.drop(part[part['Q39'] == 5].index)
part = part.drop(part[part['Q39'] == 98].index)
```

```python
part[['Q39_2', 'Q39_3', 'Q39_4']] = pd.get_dummies(part['Q39'], prefix = 'Q39', drop_first=True)
part.drop(['Q39'], axis=1, inplace=True)
```

**Q40 그대로 사용**

### Q41 -5~5 사이값으로 조정

```python
part.loc[(part['Q41'] == 1) | (part['Q41'] == 2) | (part['Q41'] == 3) | (part['Q41'] == 4) | (part['Q41'] == 5), 'Q41'] = part.loc[(part['Q41'] == 1) | (part['Q41'] == 2) | (part['Q41'] == 3) | (part['Q41'] == 4) | (part['Q41'] == 5), 'Q41'] * (-1)
part.loc[(part['Q41'] == 6) | (part['Q41'] == 7) | (part['Q41'] == 8) | (part['Q41'] == 9) | (part['Q41'] == 10), 'Q41'] = part.loc[(part['Q41'] == 6) | (part['Q41'] == 7) | (part['Q41'] == 8) | (part['Q41'] == 9) | (part['Q41'] == 10), 'Q41'] - 5
```

### Q42, Q42A 제거

```python
part.drop(['Q42', 'Q42A', 'Q42A1'], axis=1, inplace=True)
```

### Q42B 범주화

```python
sum(part['Q42B']==12)
```

```python
part = part.drop(part[part['Q42B'] == 12].index)
```

```python
part[part['Q42B'] == 1].index
```

```python
part.loc[(part['Q42B'] == 1) | (part['Q42B'] == 2) | (part['Q42B'] == 3) | (part['Q42B'] == 5) | (part['Q42B'] == 6) | (part['Q42B'] == 7), 'Q42B'] = 'a'
```

```python
part.loc[(part['Q42B'] == 4) | (part['Q42B'] == 8) | (part['Q42B'] == 9) | (part['Q42B'] == 10) | (part['Q42B'] == 11), 'Q42B'] = 'b'
```

```python
part.loc[part['Q42B'].isnull(), 'Q42B'] = 'c'
```

a : 저비용 / b : 고비용 / c : 통근, 통학 X

```python
part[['Q42B_b', 'Q42B_c']] = pd.get_dummies(part['Q42B'], prefix = 'Q42B', drop_first=True)
part.drop(['Q42B'], axis=1, inplace=True)
```

### Q42C1, Q42C2 통합(시간, 분)

```python
part.loc[part['Q42C1'].isnull(), 'Q42C1'] = 0
part.loc[part['Q42C2'].isnull(), 'Q42C2'] = 0
```

# Q44 귀하는 현재 자전거를 이용하고, 만족스러우신가요?

```python
# 과연 자전거를 어떤 수단으로 이동하고 있느냐가, 이용환경에대한 만족도에 대해 영향을 끼치고 있을까?
```

```python
plt.figure(figsize=(12,6))
plt.subplot(221)
sns.countplot(data.loc[data['Q44'] == 2,'Q44A'])
plt.title('for sports')

plt.subplot(222)
sns.countplot(data.loc[data['Q44'] == 1,'Q44A'])
plt.title('for moving')
```

```python
# 아하! 어떤 수단으로 이동하던지간에 만족도는 거의 비슷한 분포를 따름을 볼 수 있다. 
# 즉 자전거 이용 행태를 굳이 구분지어서 보지 않도록 하자. 즉. 이용함/이용하지않음 으로 구분하여보자.
# 그러면 이제 문제 44번과 문제 44-1 을 합쳐서 범주형으로 나타내어 보자.
# 자전거 이용 / 만족 , 자전거 이용/ 보통 , 자전거이용/ 불만족, 자전거 미이용 4가지 범주로 나누어 보자.
```

```python
hans['Q44A'] = data['Q44A']
# 자전거를 이용하지 않으면 Q44A 에서 nan 으로 나타날 것이다.
hans.loc[(data['Q44A'].isnull() ),'Q44A'] = 0 # 이용한함 0 
hans.loc[(data['Q44A'] == 1) | (data['Q44A'] == 2) ,'Q44A'] = 1 # 불만족 1
hans.loc[(data['Q44A'] == 3) ,'Q44A'] = 2  # 보통 2
hans.loc[(data['Q44A'] == 4) | (data['Q44A'] == 5) ,'Q44A'] = 3 #만족 3
hans = pd.get_dummies(hans, columns=['Q44A'],drop_first=True)
hans.head() 
```

# Q45 서울의 교통수단에 대한 만족도

```python
data.loc[:,'Q45A1':'Q45A3'].isnull().sum() # 이용하지 않는사람은 역시 없었다.
```

```python
hans['Q45'] = np.sum(data.loc[:,'Q45A1':'Q45A3'],axis=1)/3
```

# Q46 서울의 버스 서비스 에대한 만족도

# Q47 직업관련 질문

```python
data['Q47'].value_counts() 
# 우선 무직의 비율도 아주 많다. 무시할수는 없을듯.
```

```python
# 그냥 하면 범주가 너무 많다.
# 기타도 매우 많다... 
# 47-1 ~ 47-2 이 모두 범주로 분류되야되기 때문에 범주를 최대한 줄여보자
# 47-1 에서 성격이 비슷한 1,2/3/4,5,6 으로 분류하도록 하자. (기타는 사실 3밖에 안되서 고려 덜 해도 된다.)
data['Q47A'].value_counts() 
data.groupby(['Q47A']).aggregate({'Q47B':'mean'}) 
# 이 떄 보면 고용주/단독자영업자/상용근로자는 거의 전일제(1) 의 비율이 압도적임을 볼 수 있다.
# 즉 우리의 상식과 별반 다르지 않은 결과이므로.. 고용형태로 그냥 퉁치는게 나을 듯 하다.
# 직업에 대한 만족도는 중요한 변수이므로 범주형으로 같이 하는것보다는 따로 뺴는게 좋을듯.
```

```python
hans['Q47'] = data['Q47']
hans.loc[(data['Q47A'] == 1) | (data['Q47A'] == 2) ,'Q47'] = 1 # 고용주/ 단독자영업 
hans.loc[(data['Q47A'] == 3) ,'Q47'] = 2  # 상용근로자 2
hans.loc[(data['Q47A'] >= 4) ,'Q47'] = 3 # 무급/입시직/기타 3
hans.loc[(data['Q47']) ==2 , 'Q47' ] = 0 # 무직 =0
hans['Q47'].head()
hans = pd.get_dummies(hans, columns=['Q47'],drop_first=True)
hans
```
